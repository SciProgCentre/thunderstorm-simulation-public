```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_hist_plot(filename, energy_indx = 599):
    with open(filename, "rb") as fin:
        data = np.frombuffer(fin.read(), "d")
        bins = np.arange(0, 1001)/1000
        hist, _  = np.histogram(data, bins=bins)
        plt.subplots(nrows=1, ncols=2, figsize=(10,5))
        plt.subplot(121)
        plt.step(bins[:energy_indx], hist[:energy_indx])
        plt.xlabel("Energy, MeV")
        plt.subplot(122)
        plt.yscale("log")
        plt.step(bins[:-1], hist)
        plt.xlabel("Energy, MeV")
        return data, hist, bins
        
```

```python
read_hist_plot("cmake-build-debug/deposit.bin");
```
